```python
with open('inputs/5th_exercise_input.txt','r') as f:
    inputs = f.read().split('\n\n')
    
print(inputs)
```

```python
rules = [(item.split('|')[0], item.split('|')[1]) for item in inputs[0].split('\n')]
page_orders = [item.split(',') for item in inputs[1].split('\n')]

print(rules)
print(page_orders)
```

```python
page_priors = dict()

for rule in rules:
    if rule[0] not in page_priors:
        page_priors[rule[0]] = []
    page_priors[rule[0]].append(rule[1])

print(page_priors)
```

```python
def find_correct_orders(page_orders, page_priors):
    correct_orders = []
    incrrct_orders = []
    violation = False
    
    for page_order in page_orders:

        violation = False
        for i in range(len(page_order) - 1 ):
            if not page_order[i] in page_priors:
                violation = True
                break

            if not set(page_order[i+1:]).issubset(set(page_priors[page_order[i]])):
                violation = True
                break
        if not violation:
            correct_orders.append(page_order)
        else:
            incrrct_orders.append(page_order)
        
    return correct_orders, incrrct_orders 
```

```python
def sum_of_middle_elements(page_orders):
    sum = 0
    for page_order in page_orders:
        mid = len(page_order) // 2
        sum += int(page_order[mid])
    return sum
```

```python
def solve_incorrect_orders(incrrct_orders, page_priors):
    solved_orders = []
    for page_order in incrrct_orders:
        ranks = dict()
    
        for item in page_order:
            if not item in page_priors:
                ranks[item] = 0
                continue
            rank = len(set(page_priors[item]).intersection(set(page_order)))
            ranks[item] = rank
        
        sorted_ranks = sorted(ranks.items(), key=lambda item: item[1], reverse=True)
        solved_orders.append([item[0] for item in sorted_ranks])
        
    return solved_orders
        
        
```

```python
corect_orders , incrrct_orders = find_correct_orders(page_orders, page_priors)
print('correct_orders', corect_orders)
print('sum of middle elements', sum_of_middle_elements(corect_orders))

print('incrrct orders', incrrct_orders)

solved_incorrect_orders = solve_incorrect_orders(incrrct_orders, page_priors)
print('solved incrrct orders', solved_incorrect_orders)
print('sum of middle elements', sum_of_middle_elements(solved_incorrect_orders))
```
