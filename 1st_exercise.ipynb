```python
import numpy as np
```

```python
def total_distance(left_list, right_list):

  left_list = np.sort(left_list)
  right_list = np.sort(right_list)
  return np.sum(np.abs(left_list - right_list))
```

```python
with open('inputs/1st_exercise_input.txt', 'r') as f:
    data = f.read().splitlines()
    data = [i.split() for i in data]
    data = np.array(data, dtype=int)
    leff_list = data[:,0]
    right_list = data[:,1]

print('left_list\n',leff_list)
print('right_list\n', right_list)

print('Toatal distance between left_list and right_list\n',total_distance(leff_list, right_list))
```

```python
from collections import Counter

def calc_similarity(left_list, right_list):
    right_list_counter = Counter(right_list)
    similarity = 0
    for i in left_list:
        similarity += right_list_counter[i]*i
    return similarity
```

```python
similarity_score = calc_similarity(leff_list, right_list)
print('Similarity score between left_list and right_list\n', similarity_score)
```
