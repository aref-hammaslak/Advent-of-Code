```python
import numpy as np
```

```python
class Count_Word_Appreciate():
  def __init__(self, corpus, word):
    self.matrix =  np.array([list(row) for row in corpus.split('\n')])
    self.word = word
    self.word_reversed= word[::-1]
    self.dim = self.matrix.shape[0]

  def __call__(self):
    return self.horizontal_count() + self.vertical_count() + self.main_diagonals_count() + self.anti_diagonals_count()

  def count_report(self):
    print('Horizontal count: ', self.horizontal_count())
    print('Vertical count: ', self.vertical_count())
    print('Main diagonals count: ', self.main_diagonals_count())
    print('Anti diagonals count: ', self.anti_diagonals_count())
    print('Total count: ', self.__call__())
    
  def horizontal_count(self):

    #iterate over rows
    count = 0
    for row in self.matrix:
      for i in range(len(row) - len(self.word) + 1):
        #check the sequense match with word
        if ''.join(row[i:i+len(self.word)]) == self.word:
          count += 1
        if ''.join(row[i:i+len(self.word_reversed)]) == self.word_reversed:
          count += 1
    return count

  def vertical_count(self):
    count = 0
    #iterate over columns
    for j in range(self.dim):
      
      for i in range(self.dim - len(self.word) + 1):
        #check the sequense match with word
        if ''.join(self.matrix[i:i+len(self.word), j]) == self.word:
          count += 1
        if ''.join(self.matrix[i:i+len(self.word_reversed), j]) == self.word_reversed:
          count += 1
    return count


  def main_diagonals_count(self):

    count = 0
    #iterate over main diagonals and diagonals below main diagonal
    for t in range(self.dim-1, -1, -1):
      i = t
      j = 0
      k_word = 0
      k_word_reversed = 0
      # print('t: ' ,t)
      for _ in range(self.dim - t ):
        # print('i,j',i,j)
        #check the sequense match with word
        if self.matrix[i][j] == self.word[k_word]:
          k_word += 1
        else:
          k_word = 0

        #check the sequense match wiht word reversed
        if self.matrix[i][j] == self.word_reversed[k_word_reversed]:
          k_word_reversed += 1
        else:
          k_word_reversed = 0

        #increment the count if sequense completly match with word or word reversed
        if k_word == len(self.word) :
          count += 1
          k_word = 0
        if k_word_reversed == len(self.word_reversed):
          count += 1
          k_word_reversed = 0

        i += 1
        j += 1

    #iterate over digonals above main diagonal
    for t in range(1,self.dim):
      j = t
      i = 0
      k_word = 0
      k_word_reversed = 0
      
      # print('t:',t)
      for _ in range(self.dim - t ):
        # print('i,j',i,j)
        #check the sequense match with word
        if self.matrix[i][j] == self.word[k_word]:
          k_word += 1
        else:
          k_word = 0

        #check the sequense match wiht word reversed
        if self.matrix[i][j] == self.word_reversed[k_word_reversed]:
          k_word_reversed += 1
        else:
          k_word_reversed = 0

        #increment the count if sequense completly match with word or word reversed
        if k_word == len(self.word) :
          count += 1
          k_word = 0
        if k_word_reversed == len(self.word_reversed):
          count += 1
          k_word_reversed = 0

        i += 1
        j += 1

    return count
  def anti_diagonals_count(self):
    count = 0
    #iterate below anti diagonals and diagonals below anti diagonal
    for t in range(self.dim-1, -1, -1):
      i = t
      j = self.dim - 1
      k_word = 0
      k_word_reversed = 0
      # print('t: ' ,t)
      for _ in range(self.dim - t ):

        # print('i,j',i,j)
        #check the sequense match with word
        if self.matrix[i][j] == self.word[k_word]:
          k_word += 1
        else:
          k_word = 0

        #check the sequense match wiht word reversed
        if self.matrix[i][j] == self.word_reversed[k_word_reversed]:
          k_word_reversed += 1
        else:
          k_word_reversed = 0

        #increment the count if sequense completly match with word or word reversed
        if k_word == len(self.word) :
          count += 1
          k_word = 0
        if k_word_reversed == len(self.word_reversed):
          count += 1
          k_word_reversed = 0

        i += 1
        j -= 1

    # iterate over digonals above anti diagonal
    for t in range(0,self.dim-1):
      j = t
      i = 0
      k_word = 0
      k_word_reversed = 0
      # print('t:',t)
      for _ in range(t+1):
        # print('i,j',i,j)
        #check the sequense match with word
        if self.matrix[i][j] == self.word[k_word]:
          k_word += 1
        else:
          k_word = 0

        #check the sequense match wiht word reversed
        if self.matrix[i][j] == self.word_reversed[k_word_reversed]:
          k_word_reversed += 1
        else:
          k_word_reversed = 0

        #increment the count if sequense completly match with word or word reversed
        if k_word == len(self.word) :
          count += 1
          k_word = 0
        if k_word_reversed == len(self.word_reversed):
          count += 1
          k_word_reversed = 0

        i += 1
        j -= 1

    return count

```

```python
input = 'MMMSXXMASM\nMSAMXMSMSA\nAMXSXMAAMM\nMSAMASMSMX\nXMASAMXAMM\nXXAMMXXAMA\nSMSMSASXSS\nSAXAMASAAA\nMAMMMXMMMM\nMXMXAXMASX'
word = 'XMAS'
count_word_appreciate = Count_Word_Appreciate(input, word)
count_word_appreciate.count_report()
```
